# Pre-processing and Training Data Development

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
```

```python
tornado = pd.read_csv('/Users/gregoryolson/Documents/Data Science CT/Capstone/Capstone_Tornado/Data/tornado_downsized.csv')
```

```python
tornado.head()
```

```python
# create X and y variables
cols = tornado.columns
cols = cols.drop(['TORNADO'])

X = tornado[cols]
y = tornado['TORNADO']
```

```python
# perfom RobustScaler
transformer = RobustScaler().fit(X)
X = transformer.transform(X)
df = pd.DataFrame(X, columns=cols)
df.describe()
```

```python
# convert month columns to int
df['APRIL'] = tornado['APRIL'].astype(int)
df['JULY'] = tornado['JULY'].astype(int)
df.head()
```

```python
# assign new df to X
X = df
```

```python
# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=99)
```
